```python
import numpy  as np
import warnings
warnings.filterwarnings('ignore')
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils.np_utils import to_categorical
from keras.layers import Dropout, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D
import cv2
from sklearn.model_selection import train_test_split
import pickle
import os
import pandas as pd
import random
from keras.preprocessing.image import ImageDataGenerator
```

```python
###############Parameters####################
path="myData"
labelFile="labels.csv"
batch_size_val=50
steps_per_epochs=2000
epochs_val=5
imageDimension=(32,32,3)
testRatio=0.2
validationRatio=0.2
```

```python
count=0
images=[]
classNo=[]
myList = os.listdir(path)
print("Total Classes Detection ",len(myList))
noOfClasses = len(myList)
print("Importing Classes.....")
for x in range(0, len(myList)):
    myPicList=os.listdir(path+"/"+str(count))
    for y in myPicList:
        curImg=cv2.imread(path+"/"+str(count)+"/"+y)
        images.append(curImg)
        classNo.append(count)
    print(count)
    count+=1
print(" ")
images = np.array(images)
classNo = np.array(classNo)
```

```python
#####Split Data ############
x_train, x_test, y_train , y_test= train_test_split(images, classNo, test_size = testRatio)
x_train , x_validation , y_train, y_validation = train_test_split(x_train, y_train, test_size=validationRatio)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv("labels.csv")
num_of_samples = []
cols = 5
num_classes = noOfClasses
fig, axs = plt.subplots(nrows=num_classes, ncols = cols, figsize=(5,100))
fig.tight_layout()
for i in range(cols):
    for j,row in data.iterrows():
        x_selected=x_train[y_train == j]
        axs[j][i].imshow(x_selected[random.randint(0, len(x_selected)-1),:,:],cmap=plt.get_cmap("gray"))
        axs[j][i].axis("off")
        if i == 2:
            axs[j][i].set_title(str(j)+"-"+row["Name"])
            num_of_samples.append(len(x_selected))
            
```

```python
################Dislay Bar Chart################
print(num_of_samples)
plt.figure(figsize=(12,4))
plt.bar(range(0, num_classes), num_of_samples)
plt.title("Distribution of the training dataset")
plt.xlabel("Class Number")
plt.ylabel("Number of images")
plt.show()
```

```python
######Image Preprocessing#################
# def grayscale(img):
#     img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#     return img
# def equalize(img):
#     img=cv2.equalizeHist(img)
#     return img
# def preprocessing(img):
#     img = grayscale(img)
#     img = equalize(img)
#     img = img/255.00
#     return img

def preprocessing(img):
    img=img.astype("uint8")
    img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img=cv2.equalizeHist(img)
    img = img/255
    return img
```

```python
import numpy as np
x_train=np.array(list(map(preprocessing, x_train)))
x_validation=np.array(list(map(preprocessing, x_validation)))
x_test=np.array(list(map(preprocessing, x_test)))
```

```python
x_train=x_train.reshape(x_train.shape[0], x_train.shape[1], x_train.shape[2],1)
x_validation=x_validation.reshape(x_validation.shape[0], x_validation.shape[1], x_validation.shape[2],1)
x_test=x_test.reshape(x_test.shape[0], x_test.shape[1], x_test.shape[2],1)
```

```python
dataGen = ImageDataGenerator(
width_shift_range=0.1,
height_shift_range=0.1,
zoom_range=0.2,
shear_range=0.1,
rotation_range=10
)
```

```python
dataGen.fit(x_train)
```

```python
batches=dataGen.flow(x_train, y_train, batch_size=20)
x_batch, y_batch = next(batches)
```

```python
y_train = to_categorical(y_train, noOfClasses)
y_validation = to_categorical(y_validation, noOfClasses)
y_test = to_categorical(y_test, noOfClasses)
```

```python
def myModel():
    no_of_Filters=50
    size_of_Filter=(5,5)
    size_of_Filter2=(3,3)
    size_of_pool = (2,2)
    no_of_Nodes=500
    model=Sequential()
    model.add((Conv2D(no_of_Filters, size_of_Filter, input_shape=(imageDimension[0],imageDimension[1],1),activation='relu')))
    model.add((Conv2D(no_of_Filters, size_of_Filter,activation='relu')))
    model.add(MaxPooling2D(pool_size=size_of_pool))
    
    model.add((Conv2D(no_of_Filters//2, size_of_Filter2,activation='relu')))
    model.add((Conv2D(no_of_Filters//2, size_of_Filter2,activation='relu')))
    model.add(MaxPooling2D(pool_size=size_of_pool))
    model.add(Dropout(0.5))
              
    model.add(Flatten())
    model.add(Dense(no_of_Nodes))
    model.add(Dropout(0.5))
    model.add(Dense(noOfClasses, activation='softmax'))
    model.compile(Adam(lr=0.001),loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
model=myModel()
```

```python
model.summary()
```

```python
history=model.fit_generator(dataGen.flow(x_train, y_train, batch_size=batch_size_val),steps_per_epoch=steps_per_epochs,epochs=epochs_val, validation_data=(x_test, y_test))
```

```python
plt.figure(1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['training', 'validation'])
plt.title('loss')
plt.xlabel('epochs')
plt.figure(2)
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.legend(['training', 'validation'])
plt.title('Accuracy')
plt.xlabel('epochs')
plt.show()
```

```python
score=model.evaluate(x_test, y_test, verbose=0)
print("Test Score",score[0])
print("Test Accuracy",score[1])
```
